```python
from qiskit import QuantumCircuit
```

```python
circ = QuantumCircuit(4,2)
circ.x([0,1])
circ.barrier()
circ.cx(0,2)
circ.cx(1,2)
circ.barrier()
circ.ccx(0,1,3)
circ.measure(2,0)
circ.measure(3,1)
circ.draw(output='mpl')
```

```python
from qiskit import execute, Aer
backend = Aer.get_backend('qasm_simulator')
result = execute(circ,backend).result().get_counts()
print(result)
```
