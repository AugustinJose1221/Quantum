```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import Aer, execute
from qiskit.tools.visualization import plot_histogram, plot_state_city
```

```python
from qiskit.providers.aer import QasmSimulator, StatevectorSimulator, UnitarySimulator
```

```python
#Two qbit bell state
qr = QuantumRegister(2,'qr')
cr = ClassicalRegister(2,'cr')
circ = QuantumCircuit(qr,cr)
circ.h(qr[0]) #superposition
circ.cx(qr[0],qr[1]) #entanglement
circ.measure(qr,cr)
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circ,simulator,shots=50).result()
counts = result.get_counts(circ)
plot_histogram(counts,title='Bell-State counts')
```

```python
#Circuit initialised in custom state
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circ = QuantumCircuit(qr,cr)
circ.initialize([1,0,0,1]/np.sqrt(2),[qr[0],qr[1]])
circ.measure(qr,cr)

simulator = Aer.get_backend('qasm_simulator')

result = execute(circ,simulator).result()
counts = result.get_counts(circ)
plot_histogram(counts,title="Bell initial statevector")
```

```python
#without measurement
qr = QuantumRegister(2,'qr')
circ = QuantumCircuit(qr)
circ.h(qr[0])
circ.cx(qr[0],qr[1])

simulator = Aer.get_backend('statevector_simulator')

result = execute(circ,simulator).result()
statevector = result.get_statevector(circ)
plot_state_city(statevector,title="Bell State")
```

```python
#with measurement
qr = QuantumRegister(2,'qr')
cr = ClassicalRegister(2,'cr')
circ = QuantumCircuit(qr,cr)
circ.h(qr[0])
circ.cx(qr[0],qr[1])
circ.measure(qr,cr)

simulator = Aer.get_backend('statevector_simulator')

result = execute(circ,simulator).result()
statevector = result.get_statevector(circ)
plot_state_city(statevector,title='Bell state after measurement')
```

```python
qr = QuantumRegister(2)
circ = QuantumCircuit(qr)
circ.h(qr[0])
circ.cx(qr[0],qr[1])

simulator = Aer.get_backend('unitary_simulator')

result = execute(circ,simulator).result()
unitary = result.get_unitary(circ)
print("Circuit Unitary: \n",unitary)
```
